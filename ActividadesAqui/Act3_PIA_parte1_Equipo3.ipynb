```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

ruta_numero="/content/numeros"
numeros_numeros=[]
img_size=28

for img in os.listdir(ruta_numero) :
 img = cv2.imread(os.path.join(ruta_numero,img))
 img_gray = cv2.cvtColor(img,cv2.IMREAD_GRAYSCALE)
 img_gray_resize = cv2.resize(img_gray,(img_size,img_size))
 numeros_numeros.append([img_gray_resize])

print(len(numeros_numeros))
print(numeros_numeros)
```

```python
plt.figure()
plt.imshow(np.squeeze(numeros_numeros[264]))
plt.colorbar
plt.grid(False)
plt.show()
```
